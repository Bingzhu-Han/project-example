```python
# define a function to repeat the process to all the boroughs in London
def get_near_venues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
        url='https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID,CLIENT_SECRET,lat,lng,VERSION,radius,LIMIT)
        results = requests.get(url).json()['response']['groups'][0]['items']
    
        venues_list.append([(
        name,
        lat,
        lng,
        v['venue']['name'],
        v['venue']['location']['lat'],
        v['venue']['location']['lng'],
        v['venue']['categories'][0]['name']) for v in results])
    
    near_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    near_venues.columns=['Borough','Borough Latitude','Borough Longitude','Venue','Venue Latitude','Venue Longitude','Venue Category']
    
    return(near_venues)
```

```python
london_venues = get_near_venues(names=df['Borough'],latitudes=df['Latitude'],longitudes=df['Longitude'])
```

```python
# check the result
london_venues.head()
```

```python
# explore each borough in London
# one hot coding
london_onehot = pd.get_dummies(london_venues[['Venue Category']], prefix="", prefix_sep="")
london_onehot['Borough'] = london_venues['Borough'] 
fixed_columns = [london_onehot.columns[-1]] + list(london_onehot.columns[:-1])
london_onehot = london_onehot[fixed_columns]

london_onehot.head()
```

```python
# group the boroughs and take frequency of each category
london_grouped = london_onehot.groupby(['Borough'],as_index=False).mean()
london_grouped
```

```python
# find the top 5 venues within each borough
num_top_venues = 5

for bor in london_grouped['Borough']:
    print("----"+bor+"----")
    temp = london_grouped[london_grouped['Borough'] == bor].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# define a function to sort the venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# display top 10 venues of each borough
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
borough_venues_sorted = pd.DataFrame(columns=columns)
borough_venues_sorted['Borough'] = london_grouped['Borough']

for ind in np.arange(london_grouped.shape[0]):
    borough_venues_sorted.iloc[ind, 1:] = return_most_common_venues(london_grouped.iloc[ind, :], num_top_venues)

borough_venues_sorted
```

We can see that Chinese restaurant is the 2nd most common venue in Greenwich. As a thumb of rule, "Economy of Scale" tells us that we should open a Chinese restaurant in Greenwich.

```python
# cluster the boroughs
london_cluster = london_grouped.drop('Borough',axis=1)
# find the best k
for k in range(1,10):
    kmeans=KMeans(n_clusters=k,random_state=0).fit(london_cluster.iloc[:,:])
    labels=kmeans.labels_
    dis=kmeans.inertia_
    print('k',k,'cost:',dis)
```

```python
# cluster boroughs using K-Means
k=3
kmeans = KMeans(n_clusters=k, random_state=0).fit(london_cluster)
kmeans.labels_
```

```python
# create a new dataframe to present the boroughs with clusters
borough_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
london_merged = df

london_merged = london_merged.join(borough_venues_sorted.set_index('Borough'), on='Borough')

london_merged
```
