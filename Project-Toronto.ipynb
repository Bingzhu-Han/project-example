```python
def get_cat_type(row):
    try:
        cat_list = row['categories']
    except:
        cat_list = row['venue.categories']
    
    if len(cat_list)==0:
        return None
    else:
        return cat_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
near_venues = json_normalize(venues)
f_columns = ['venue.name','venue.categories','venue.location.lat','venue.location.lng']
near_venues = near_venues.loc[:,f_columns]
near_venues['venue.categories'] = near_venues.apply(get_cat_type, axis=1)
near_venues.columns = [col.split('.')[-1] for col in near_venues.columns]
near_venues.head()
```

Let's create a function to repeat the process to all the neighbourhoods in selected boroughs.

```python
def getNearVenues(names,latitudes,longitudes,radius=500):
    
    venue_list=[]
    for name,lat,lng in zip(names, latitudes, longitudes):
        print(name)
        
        url='https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID,CLIENT_SECRET,lat,lng,VERSION,radius,LIMIT)
        
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        venue_list.append([(name,
                            lat,
                            lng,
                            v['venue']['name'],
                           v['venue']['location']['lat'],
                           v['venue']['location']['lng'],
                           v['venue']['categories'][0]['name']) for v in results])
        
    near_venues = pd.DataFrame([item for venue_list in venue_list for item in venue_list])
        
    near_venues.columns = ['Neighbourhood',
                              'Neighbourhood Latitude',
                              'Neighbourhood Longitude',
                              'Venue',
                              'Venue Latitude',
                              'Venue Longitude',
                              'Venue Category']
        
    return(near_venues)
```

```python
t_venues = getNearVenues(names=tdf['Neighbourhood'],
                        latitudes=tdf['Latitude'],
                        longitudes=tdf['Longitude'])
```

Now let's analyze each neighbourhood.

```python
# one hot coding
t_onehot = pd.get_dummies(t_venues[['Venue Category']],prefix='',prefix_sep='')
t_onehot['Neighbourhood'] = t_venues['Neighbourhood']
fixed_columns = [t_onehot.columns[-1]]+list(t_onehot.columns[:-1])
t_onehot = t_onehot[fixed_columns]
t_onehot.head()
```

```python
t_onehot.shape
```

```python
t_grouped = t_onehot.groupby('Neighbourhood').mean().reset_index()
t_grouped
```

Let's print each neighbourhood along with top 5 most common venues.

```python
num_top_venues = 5

for hood in t_grouped['Neighbourhood']:
    print('----'+hood+'----')
    temp = t_grouped[t_grouped['Neighbourhood']== hood].T.reset_index()
    temp.columns=['venue','freq']
    temp=temp.iloc[1:]
    temp['freq']=temp['freq'].astype(float)
    temp=temp.round({'freq':2})
    print(temp.sort_values('freq',ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# use a function to sort these venues in descending order
def return_sort_venues(row,num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
# create a new dataframe and display top 10 venues for each neighbourhood
num_top_venues=10
indicators = ['st','nd','rd']

columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1,indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
        
n_sorted = pd.DataFrame(columns=columns)
n_sorted['Neighbourhood'] = t_grouped['Neighbourhood']

for ind in np.arange(t_grouped.shape[0]):
    n_sorted.iloc[ind,1:] = return_sort_venues(t_grouped.iloc[ind,:],num_top_venues)
    
n_sorted.head()
```

Run k-means to cluster these neighbourhood into 5 clusters.

```python
from sklearn.cluster import KMeans
k=5
t_cluster = t_grouped.drop('Neighbourhood',1)
kmeans = KMeans(n_clusters=k,random_state=0).fit(t_cluster)
kmeans.labels_[0:10]
```

```python
# create a dataframe to display top 10 venues for each neighbourhood
n_sorted.insert(0,'Cluster Labels',kmeans.labels_)

t_merged = tdf

t_merged = t_merged.join(n_sorted.set_index('Neighbourhood'),on='Neighbourhood')

t_merged.head()
```

```python
# visualize
# create a map
map_clusters = folium.Map(location=[latitude,longitude],zoom_start=11)
# set color schemes
x=np.arange(k)
ys=[i+x+(i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0,1,len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
# add markers to map
markers_color = []
for lat,lng,poi,cluster in zip(t_merged['Latitude'],t_merged['Longitude'],t_merged['Neighbourhood'],t_merged['Cluster Labels']):
    label = folium.Popup(str(poi)+'Cluster'+str(cluster),parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color=rainbow[cluster-1],
    fill=True,
    fill_color=rainbow[cluster-1],
    fill_opacity=0.7).add_to(map_clusters)
    
map_clusters
```
